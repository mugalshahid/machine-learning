```python
import numpy as np
import pandas as pd
```

```python
# Encoding = Transform your categorical data into numerical data
# Normal Distribution convert
```

```python
df = pd.read_csv("covid_toy - covid_toy.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy = 'mean')
df['fever'] = si.fit_transform(df[['fever']])
```

```python
df.head(2)
```

```python
# Encoding  =  Transform data into numerical data
# gender = M , F ---> 0 , 1
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()

df['gender'] = lb.fit_transform(df['gender'])
df['cough'] = lb.fit_transform(df['cough'])
df['city'] = lb.fit_transform(df['city'])
df['has_covid'] = lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
np.round(x_train.describe() , 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)  ## fit means learn the parameters and transform means apply on that data
```

```python
x_train_new = pd.DataFrame(x_train_sc , columns = x_train.columns)
```

```python
np.round(x_train_new.describe(), 2)
```

# Normalization = min_val : 0 and max_val : 1

```python
np.round(x_train.describe(),2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn , columns = x_train.columns)
```

```python
np.round(x_train_new.describe(),2)
```
